```python
# Importing required packages and loading dataset to pandas dataframe
# And renaming columns to appropriate name
import pandas as pd
column_names = ['uuid','price_string','price_unfiltered',\
                'product_type','level_1','category']
df = pd.read_csv('test.csv', names = column_names, skiprows =1)
df.head()
```

```python
# Checking for null/missing values in data for each column
null_values = df.isna().sum()
print('column name          null values')
print('-------------------------------------')
print(null_values)
```

```python
# Ans1 Products details without their price in dataframe ans1
# Took all the columns except price details columns
ans1 = df[['uuid','product_type','category','level_1']]
print("The dataframe ans1 shown below contains product "
    "details excluding their price details.")
ans1.head()
```

```python
# Ans2a Count of products without prices in data
# Taking product and price in a temporary dataframe
tempdf = df[['product_type','price_string']]

# Taking rows with not null value for product_type
tempdf = tempdf[tempdf['product_type'].notna()]

# Taking rows with null value for price
tempdf = tempdf[tempdf['price_string'].isna()]

# Grouping on product and taking count of each with product with
# no price value/missing price value
count = tempdf.groupby('product_type').size()

# Taking count of products with no price value/missing price value
print('Count of products with no price information is', count.count())
```

```python
"""
Ans2b Count of products with prices in each product_type, 
category and level_1
"""

tempdf2 = df.dropna() # removing rows with null values

"""
Grouping products with non null price value in each product,
category and level_1, and taking their count
"""

ans2b = tempdf2.groupby(['product_type','category','level_1'])\
    .size().reset_index()

print("The table below shows count of products with price values"
      " availabe in dataset for each product_type, category, level_1")

# Renaming the output dataframe count2
ans2b.columns = ['product_type', 'category', 'level_1', 'count']
ans2b.head()
```

```python
# Ans3a Adding $ sign to price_string
tempdf3 = df # taking dataframe into temporary variable

# Stripping $ sign if present in price_string
tempdf3['new_price_string'] = tempdf3['price_string'].str.replace('$', '')

# Adding $ sign to price values
tempdf3['new_price_string'] = '$'+ tempdf3['new_price_string']
# Taking relevant columns 

print('Below is the dataframe conveting price values to correct form : $price')
ans3a = tempdf3[['uuid','new_price_string','product_type','category','level_1']]
ans3a.head(10)
```

```python
# Ans3b Seperating currency and price values

import pandas as pd
# Disabling chained assignment warning
pd.options.mode.chained_assignment = None  
ans3a['new_price_string'] = ans3a['new_price_string'].str.replace('$', '')
ans3a.rename(columns = {'new_price_string':'price_value'}, inplace = True)

# Adding a currency column with value as $
ans3a['currency'] = '$'

# Rearranging dataframe with relevant columns
ans3b = ans3a[['uuid','currency','price_value','product_type',\
                   'category','level_1']]

print('The dataframe below has currency and price values in seperate'
     ' columns')
ans3b.head(10)
```

```python
# Ans4 Average price for products per category

# Taking dataframe of last result in temporary dataframe
# For using in next question
tempdf4 = ans3b

# Convertin price value to float
tempdf4['price_value'] = tempdf4['price_value'].astype(float)
tempdf4 = tempdf4[['category','price_value']]

# Grouping over category and taking average price of product for each category
ans4 = tempdf4.groupby(['category'])['price_value'].mean().reset_index()
ans4.rename(columns = {"price_value": "average_price_of_product"},\
            inplace = True)
ans4['currency'] = '$' # denoting currency for average price of product 
print('The dataframe below shows the average price of product'
     ' per category')
# Adding the currency column
ans4 = ans4[['category','currency','average_price_of_product']]
ans4.head(10)
```
